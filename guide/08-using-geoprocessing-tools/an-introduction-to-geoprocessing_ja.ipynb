# ジオプロセシング

ArcGIS を使用して、地理情報を解析するだけでなく、地理データを用いたマッピングやその探索を行うこともできます。これは、空間解析ツールを使用して既存のデータセットから新しい情報を導き出すために使います。例えば、ポイントのセットを指定すると、1 マイルのバッファー ゾーンを計算し、これを使用して他の情報を選択できます。

ArcGIS では、ジオプロセシング ツールを使用して、これらのタスクを実行します。ジオプロセシング ツールは、地理データに対して操作を実行する機能です。一般的なジオプロセシング操作では、入力データセットを取得し、そのデータセットに対して操作を実行し、操作の結果を出力データセットとして返します。ジオプロセシングの一般的な操作には、ジオグラフィック フィーチャのオーバーレイ、フィーチャの選択と解析、トポロジー処理、ラスター処理、データ変換などがあります。ジオプロセシングでは、意思決定を行うために使用する情報の定義、管理、および分析が可能になります。

## ArcGIS API for Python を用いたジオプロセシング

ArcGIS API for Python は、ArcGIS で実行できるジオプロセシング機能の Python プログラミング インターフェイスを提供します。ジオプロセシングは GIS の不可欠な部分であり、Python API はモジュールの設計にこの概念を反映しています。arcgis パッケージはモジュールに編成され、GIS のさまざまな空間解析機能または地理データセットに対する科学的プログラミング インターフェイスを提供します。以下にはその関数の例を示します。

* `arcgis.features` : GIS 内のフィーチャ データ、フィーチャ レイヤー、およびフィーチャ レイヤー コレクションを操作するために使用されます。また、フィーチャ データを操作する空間解析ジオプロセシング ツールも含まれています。

* `arcgis.raster` : ラスター データとイメージ レイヤーを操作するためのクラスやラスター解析のジオプロセシング ツールが含まれています。

* `arcgis.network` : 最寄り施設の検出、車両による移動時の最適ルート、ロケーション アロケーション解析、OD コスト マトリックス解析、到達圏の生成等、ネットワーク分析のためのクラスと関数が含まれています。

* `arcgis.geoanalytics` : 大規模なデータセットの分散解析用のデータタイプとジオプロセシング ツールです。

ジオプロセシング ツールは API の関数として表され、適用先の空間データの種類を表すモジュール内に配置されます。開発者は、プログラム内のモジュールをインポートし、その中で定義されたデータタイプと関数を使用してジオプロセシングを実行します。

```
Note : ArcGIS API for Python では、Esri によって構築された豊富なジオプロセシング ツールが含まれています。API に組み込まれたジオプロセシング ツールのみを使用する場合は、作業している空間データの種類に関するドキュメントを参照してください。
```
